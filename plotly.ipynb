# How to Calculate Batting Average in Cricket?

Calculating the batting average of a player needs a few values. The first would be the total number of runs the player has scored. Second, the number of innings that the player has batted in and third, the number of times that the player has been dismissed or the number of times that the player has remained not out.

The batting average isnâ€™t the number of runs divided by the number of matches or the number of innings played. 
### To calculate the batting average of any player, you need to divide the total number of runs scored by that player by the number of times he or she has got out (or been dismissed). We can calculate the number of times the player has been dismissed by subtracting the number of not outs from the number of innings batted.

So, for instance, if a player, say Player A, plays 300 matches but bats in only 250 innings and is dismissed only in 200 of those, then the average is calculated using the 200 as the denominator and not 300 or 250. So, if Player A has made 10000 runs in his playing career, then his average would be calculated by dividing 10000 by 200. Therefore, the average of Player A would be 50.

```python
# year by year performance

single = ipl[ipl.batsman=='V Kohli']
performance = single.groupby('season')['batsman_runs'].sum().reset_index()
performance
```

```python
# plot scatter
trace = go.Scatter(x=performance.season,
                   y=performance.batsman_runs,
                   mode='lines',marker={'color':'#00a65a'})

data = [trace]

layout = go.Layout(title='Year by Year Performnace',
                  xaxis={'title':'Season'},
                  yaxis={'title':'Runs'})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='year by year performance')
```

```python
# year by year batsman performance

single = ipl[ipl.batsman=='V Kohli']
performance = single.groupby('season')['batsman_runs'].sum().reset_index()
performance

single1 = ipl[ipl.batsman=='MS Dhoni']
performance1 = single1.groupby('season')['batsman_runs'].sum().reset_index()
performance1
```

```python
combine = performance.merge(performance1,on='season')
combine
```

```python
trace = go.Scatter(x=performance.season,
                   y=performance.batsman_runs,
                   mode='lines+markers',marker={'color':'#00a65a'}
                  ,name = 'Virat Kohli')

trace1 = go.Scatter(x=performance1.season,
                   y=performance1.batsman_runs,
                   mode='lines+markers',
                   name='MS Dhoni')

data = [trace,trace1]

layout = go.Layout(title='Year by Year Performnace',
                  xaxis={'title':'Season'},
                  yaxis={'title':'Runs'})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='year by year performance combine')
```

```python
# Multiple Line Charts

def batsman_comp(*name):
    data = []
    
    for i in name:
        single = ipl[ipl.batsman==i]
        performance = single.groupby('season')['batsman_runs'].sum().reset_index()
        
        trace = go.Scatter(x=performance.season,
                   y=performance.batsman_runs,
                   mode='lines+markers',
                   name=i)
        data.append(trace)
        
    layout = go.Layout(title='Batsman record Comparator',
                  xaxis={'title':'Season'},
                  yaxis={'title':'Runs'})
    
    fig = go.Figure(data=data,layout=layout)
    
    pyo.plot(fig,filename='year_by_year1')
    
```

```python
batsman_comp('V Kohli', 'RG Sharma', 'DA Warner', 'MS Dhoni')
```

# 3. Bar PLot

```python
top10= ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).index.to_list()
```

```python
top10_df = ipl[ipl['batsman'].isin(top10)]
top10_df
```

```python
top10_score = top10_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).reset_index()
top10_score
```

```python
# bar plot
trace = go.Bar(x=top10_score.batsman,y=top10_score.batsman_runs)

data = [trace]

layout = go.Layout(title='Top10 batsman score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Batsman_runs'})

fig = go.Figure(data=data,layout=layout)
    
pyo.plot(fig,filename='Top10 batsman score')
```

```python
final
```

```python
# Nested Bar Graph
trace1 = go.Bar(x=final.batsman,y=final['1st innings'], 
                name = '1st innings',
               marker = {'color':'#00a65a'})

trace2 = go.Bar(x=final.batsman,y=final['2nd innings'], 
                name = '2nd innings',
               marker = {'color':'red'})
data = [trace1,trace2]

layout = go.Layout(title='Inning wise score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Batsman_runs'})

fig = go.Figure(data=data,layout=layout)
    
pyo.plot(fig,filename='Inning wise score')
```

```python
# Overlayed bar graph
trace1 = go.Bar(x=final.batsman,y=final['1st innings'], 
                name = '1st innings',
               marker = {'color':'#00a65a'})

trace2 = go.Bar(x=final.batsman,y=final['2nd innings'], 
                name = '2nd innings',
               marker = {'color':'red'})
data = [trace1,trace2]

layout = go.Layout(title='Inning wise score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Batsman_runs'},
                  barmode= 'overlay')

fig = go.Figure(data=data,layout=layout)
    
pyo.plot(fig,filename='Inning wise score1')
```

```python
# Stacked Bar graph
trace1 = go.Bar(x=final.batsman,y=final['1st innings'], 
                name = '1st innings',
               marker = {'color':'#00a65a'})

trace2 = go.Bar(x=final.batsman,y=final['2nd innings'], 
                name = '2nd innings',
               marker = {'color':'red'})
data = [trace1,trace2]

layout = go.Layout(title='Inning wise score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Batsman_runs'},
                  barmode= 'stack')

fig = go.Figure(data=data,layout=layout)
    
pyo.plot(fig,filename='Inning wise score2')
```
